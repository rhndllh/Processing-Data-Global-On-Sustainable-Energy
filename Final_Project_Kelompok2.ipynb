## Pre-Modeling

```python
#import library yang akan digunakan pada proses pre modeling
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Label Encoder untuk konversi data string ke numerik
LEncoder = LabelEncoder()

df['Wilayah'] = LEncoder.fit_transform(df['Wilayah'])
```

```python
df.info()
```

```python
df['Wilayah']
```

```python
#menentukan kolom untuk atribut(x) dan label(y)
X = df.drop('Wilayah',axis=1).values
y = df['Wilayah'].values
```

```python
#split data testing dan data training dengan test size 0,4 
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.4, random_state=42)
```

```python
#scaler data untuk memperkecil penyimpangan 
sc = StandardScaler()

X_train = sc.fit_transform(x_train)
X_test= sc.transform(x_test)
```

## Modeling

```python
# import library yang akan digunakan pada proses Modeling

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
```

```python
# training dan testing dengan Logistic Regression

logreg= LogisticRegression()

logreg.fit(X_train, y_train)

y_pred_logreg = logreg.predict(X_test)
```

```python
# training dan testing dengan Gradient Boosting Classifier


GB = GradientBoostingClassifier()

GB.fit(X_train, y_train)

y_pred_GB = GB.predict(X_test)
```

```python
# training dan tetsing data dengan Random Forest Classifier

rf = RandomForestClassifier()

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
```

```python
# training dan testing data dengan Decision Tree Classifier

dt = DecisionTreeClassifier()

dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
```

```python
# training dan testing data dengan KNeighbors Classifier


knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)
```

```python
# training dan testing data dengan XGB Classifier

XGB = XGBClassifier() 

XGB.fit(X_train, y_train)

y_pred_XGB = XGB.predict(X_test)
```

```python
# training dan testing data dengan Support Vector classifier

svc = SVC(probability=True)

svc.fit(X_train,y_train)

y_pred_svc = svc.predict(X_test)
```

```python
# menampilkan dimensi data training dan testing 
X_train.shape, y_train.shape,X_test.shape, y_test.shape
```

## Evaluation

```python
#import library yang akan digunakan pada proses evaluasi
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score, recall_score, confusion_matrix, precision_recall_fscore_support
from sklearn.metrics import roc_auc_score,auc,f1_score
from sklearn.metrics import precision_recall_curve,roc_curve
```

```python
#menmapilkan akurasi skor masing masing machine learning
models = []

Z = [SVC() , DecisionTreeClassifier() , LogisticRegression() , KNeighborsClassifier() ,XGBClassifier(),
    RandomForestClassifier() , GradientBoostingClassifier()]


X = ["SVC" , "DecisionTreeClassifier" , "LogisticRegression" , "KNeighborsClassifier" ,
    "RandomForestClassifier" , "GradientBoostingClassifier", "XGB"]

for i in range(0,len(Z)):
    model = Z[i]
    model.fit( X_train , y_train )
    pred = model.predict(X_test)
    models.append(accuracy_score(pred , y_test))
```

```python
d = { "Accuracy" : models , "Algorithm" : X }
data_frame = pd.DataFrame(d)
data_frame
```
